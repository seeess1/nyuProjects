## Overview

As a budding data scientist and a longtime fan of tennis, I decided to analyze access to public tennis facilities in New York City in terms of various socioeconomic factors. This analysis combines Public Use Microdata Areas (PUMA), NYC Open Data, Google Maps API geocoding, and demographic and economic American Community Survey (ACS) data. After cleaning, merging, and plotting the data, I applied various multivariate regressions to predict access to public tennis facilities. Ultimately, the best model in terms of simplicity and explanatory power was the one fitting an OLS model to accessibility based on the percentage of families living in poverty within a PUMA, showing that, by this measure, NYC actually reverses the common stereotype that only the rich have access to tennis.

Read below to see my code, plots, and regressions. Or skip ahead to my project write-up, which you can find here: https://www.authorea.com/users/249384/articles/347396-public-tennis-courts-in-nyc.

### Google Maps API query

Because some of the tennis facilities provided in the NYC Open Data file were missing latitude and longitude data (and therefore could not be used in the rest of my analysis), I decided to use Google Maps API geocoding to find their latitude and longitude. 

I tried to figure out how to send a batch query using the Google Maps API to geocode each public tennis facility missing latitude and longitude data. I was unable to write the script myself so, through research, I found a script published online by Irish data scientist Shane Lynn for exactly this task. I modified the script for my own purposes (mainly by taking out functions created to track the timing and volume of the query) since my query was for a very small set of data. 

Huge thanks to Shane for publishing his script and making things a lot easier. His script can be found at the following URL:

https://www.shanelynn.ie/batch-geocoding-in-python-with-google-geocoding-api/#script-code

Note that in order for the query and the next couple cells to run, you will need your own Google Maps geocoding API key saved in a text file called "google_api.txt." Do not include any other information in that file.

```python
# To retrieve API key for following query
API_KEY = open('google_api.txt', 'r').readlines()[0][:] # api stored in seperate .txt file
```

```python
# Form a list of addresses for geocoding:
addresses = (tennis_for_geocode['name'] + ',' + 'NY,US').tolist()

# Function
def get_google_results(address, api_key=None, return_full_response=False):
    # Set up Geocoding url
    geocode_url = "https://maps.googleapis.com/maps/api/geocode/json?address={}".format(address)
    if api_key is not None:
        geocode_url = geocode_url + "&key={}".format(api_key)
        
    # Ping google for the reuslts:
    results = requests.get(geocode_url)
    # Results will be in JSON format - convert to dict using requests functionality
    results = results.json()
    
    # if there's no results or an error, return empty results.
    if len(results['results']) == 0:
        output = {
            "formatted_address" : None,
            "latitude": None,
            "longitude": None,
            "accuracy": None,
            "google_place_id": None,
            "type": None,
            "postcode": None
        }
    else:    
        answer = results['results'][0]
        output = {
            "formatted_address" : answer.get('formatted_address'),
            "latitude": answer.get('geometry').get('location').get('lat'),
            "longitude": answer.get('geometry').get('location').get('lng'),
            "accuracy": answer.get('geometry').get('location_type'),
            "google_place_id": answer.get("place_id"),
            "type": ",".join(answer.get('types')),
            "postcode": ",".join([x['long_name'] for x in answer.get('address_components') 
                                  if 'postal_code' in x.get('types')])
        }
        
    # Append some other details:    
    output['input_string'] = address
    output['number_of_results'] = len(results['results'])
    output['status'] = results.get('status')
    if return_full_response is True:
        output['response'] = results
    
    return output

# Create a list to hold results
results = []
# Go through each address in turn
for address in addresses:
    geocode_result = get_google_results(address, API_KEY)
    results.append(geocode_result)           

tennis_geocode_done = pd.DataFrame(results)
```

```python
tennisgdf_sjoin = gpd.sjoin(tennisgdf,pumashp)
```

```python
# Note that I use 15840 because that is equivalent to 3 miles in feet, which is the unit of measurement for this
# projection system
def court_counter(point):
    court_count = 0
    for index, row in tennisgdf_sjoin.iterrows():
        if point.distance(row.geometry) < 15840:
            court_count += row['crt_num']
    return court_count
```

```python
centroids['all_courts'] = centroids['geometry'].apply(court_counter)
```

## Demographic analysis

Using the ACS's demographic data, I created new calculated columns defining access to public tennis courts for white residents, non-white residents (total population of each PUMA minus the total number of white residents), residents who are 18 years old or younger, and residents that are 65 and older.

```python
# Find the number of courts for various attributes

# Number of non-white residents
demo_merge['nonwhite_residents_abs'] = demo_merge['Pop_2E'] - demo_merge['WtNHE']
demo_merge['nonwhite_residents_perc'] = 100 - demo_merge['WtNHP']

# Courts per non-white and white residents
demo_merge['cs_per_nonwhite'] = (demo_merge['all_courts'] / (demo_merge['Pop_2E'] - demo_merge['WtNHE']))*1e3
demo_merge['cs_per_white'] = (demo_merge['all_courts'] / demo_merge['WtNHE'])*1e3

# Courts per residents 18&under and 65&above
demo_merge['cs_per_under_18'] = (demo_merge['all_courts'] / demo_merge['PopU181E'])*1e4
demo_merge['cs_per_above_65'] = (demo_merge['all_courts'] / demo_merge['Pop65pl1E'])*1e4
```

```python
# Plot of public tennis courts for non-white and white residents
fig = plt.figure(figsize=(12,12))

# Courts per 1000 non-white
ax1 = fig.add_subplot(221)
ax1.get_xaxis().set_visible(False)
ax1.get_yaxis().set_visible(False)
demo_merge.plot(column="cs_per_nonwhite", alpha=.8, linewidth=0.5, ax=ax1, \
              scheme="quantiles", legend=True, cmap=plt.cm.get_cmap('magma'))
tennisgdf.plot(alpha=1, ax=ax1, c='red', markersize=10)
leg = ax1.get_legend()
leg.set_bbox_to_anchor((0.35, 0.95, 0, 0))
ax1.set_title("Courts Per 1000 Non-White Residents", fontsize=15)

# Courts per 1000 white
ax2 = fig.add_subplot(222)
ax2.get_xaxis().set_visible(False)
ax2.get_yaxis().set_visible(False)
demo_merge.plot(column="cs_per_white", alpha=.8, linewidth=0.5, ax=ax2, \
              scheme="quantiles", legend=True, cmap=plt.cm.get_cmap('magma'))
tennisgdf.plot(alpha=1, ax=ax2, c='red', markersize=10)
leg = ax2.get_legend()
leg.set_bbox_to_anchor((0.35, 0.95, 0, 0))
ax2.set_title("Courts Per 1000 White Residents", fontsize=15)

# Courts per 10k 18&under
ax3 = fig.add_subplot(223)
ax3.get_xaxis().set_visible(False)
ax3.get_yaxis().set_visible(False)
demo_merge.plot(column="cs_per_under_18", alpha=.8, linewidth=0.5, ax=ax3, \
              scheme="quantiles", legend=True, cmap=plt.cm.get_cmap('magma'))
tennisgdf.plot(alpha=1, ax=ax3, c='red', markersize=10)
leg = ax3.get_legend()
leg.set_bbox_to_anchor((0.35, 0.95, 0, 0))
ax3.set_title("Courts Per 10k Residents 18 & under", fontsize=15)

# Courts per 10k 65&above
ax4 = fig.add_subplot(224)
ax4.get_xaxis().set_visible(False)
ax4.get_yaxis().set_visible(False)
demo_merge.plot(column="cs_per_above_65", alpha=.8, linewidth=0.5, ax=ax4, \
              scheme="quantiles", legend=True, cmap=plt.cm.get_cmap('magma'))
tennisgdf.plot(alpha=1, ax=ax4, c='red', markersize=10)
leg = ax4.get_legend()
leg.set_bbox_to_anchor((0.35, 0.95, 0, 0))
ax4.set_title("Courts Per 10k Residents 65 & above", fontsize=15)
```

### Note on above plots

The above plots show the locations of public tennis facilities plotted on top of NYC PUMAs modified with scaled colorations based on various categories. The top two plots compare access to public tennis courts per 10,000 residents in each PUMA for white and non-white residents while the bottom two compare access per 10,000 residents under 18 years old with residents who are 65 and older. Note that access is defined as the number of courts within 3 miles of each PUMA centroid with the color gradients on a quantile scale.

Residents 65 and older appear to have slightly more access than residents under 18 years old given that most PUMAs have at least 19 courts for residents 65 and older while the majority of PUMAs have between 10 and 18 courts for residents 18 and under. 

The difference in access for non-white vs. white PUMA residents is less obvious. There is overlap in access among some PUMAs for white and non-white residents, however, the PUMAs with the greatest access are entirely different from each other. This is mostly explained by the demographic characteristics of NYC's PUMAs rather than the number of tennis courts. The PUMAs with the most courts per non-white resident also have relatively low percentages of non-white residents. So the number of courts per non-white resident increases drastically for this reason. The same phenomena, but reversed, explains the PUMAs with the greatest access for white residents.

```python
# Example showing how the areas with the most access to tennis courts for non-white residents are also areas that 
# simply have a low percentage of non-white residents.
demo_merge.sort_values(by='cs_per_nonwhite', ascending=False).head()
```

```python
econ_merge = pd.merge(econ_merge,centroids[['GeoID','all_courts']],on='GeoID', how='left')
econ_merge.fillna(0, inplace=True)
```

```python
econ_merge.columns
```

```python
# Calculations for the number of courts per type of worker

# White collar workers
econ_merge['wcw_perc'] = (econ_merge['MgBSciArtE'] + econ_merge['SalesOffE'])/econ_merge['CvEm16pl2E']
econ_merge['wcw_abs'] = econ_merge['MgBSciArtE'] + econ_merge['SalesOffE']
econ_merge['crts_per_wcw'] = (econ_merge['all_courts'] / (econ_merge['MgBSciArtE'] + econ_merge['SalesOffE']))*1e4

# Blue collar workers
econ_merge['bcw_perc'] = (econ_merge['NRCnstMntE'] + econ_merge['PrdTrnsMME'])/econ_merge['CvEm16pl2E']
econ_merge['bcw_abs'] = econ_merge['NRCnstMntE'] + econ_merge['PrdTrnsMME']
econ_merge['crts_per_bcw'] = (econ_merge['all_courts'] / (econ_merge['NRCnstMntE'] + econ_merge['PrdTrnsMME']))*1e4

# "Pink" collar workers (service industry, which is a mix of white and blue collar)
econ_merge['pcw_perc'] = econ_merge['SrvcE']/econ_merge['CvEm16pl2E']
econ_merge['pcw_abs'] = econ_merge['SrvcE']
econ_merge['crts_per_pcw'] = (econ_merge['all_courts'] / econ_merge['SrvcE'])*1e4
```

```python
# Plot of public tennis courts for non-white and white residents
fig = plt.figure(figsize=(20,20))

# Courts per 10k white collar workers
ax1 = fig.add_subplot(221)
ax1.get_xaxis().set_visible(False)
ax1.get_yaxis().set_visible(False)
econ_merge.plot(column="crts_per_wcw", alpha=.8, linewidth=0.5, ax=ax1, \
              scheme="quantiles", legend=True, cmap=plt.cm.get_cmap('magma'))
tennisgdf.plot(alpha=1, ax=ax1, c='red', markersize=10)
leg = ax1.get_legend()
leg.set_bbox_to_anchor((0.35, 0.95, 0, 0))
ax1.set_title("Courts per 10k White collar workers", fontsize=25)

# Courts per 10k blue collar workers
ax2 = fig.add_subplot(222, sharey=ax1)
ax2.get_xaxis().set_visible(False)
ax2.get_yaxis().set_visible(False)
econ_merge.plot(column="crts_per_bcw", alpha=.8, linewidth=0.5, ax=ax2, \
              scheme="quantiles", legend=True, cmap=plt.cm.get_cmap('magma'))
tennisgdf.plot(alpha=1, ax=ax2, c='red', markersize=10)
leg = ax2.get_legend()
leg.set_bbox_to_anchor((0.35, 0.95, 0, 0))
ax2.set_title("Courts per 10k Blue collar workers", fontsize=25)

# Courts per 10k pink collar workers
ax3 = fig.add_subplot(223)
ax3.get_xaxis().set_visible(False)
ax3.get_yaxis().set_visible(False)
econ_merge.plot(column="crts_per_pcw", alpha=.8, linewidth=0.5, ax=ax3, \
              scheme="quantiles", legend=True, cmap=plt.cm.get_cmap('magma'))
tennisgdf.plot(alpha=1, ax=ax3, c='red', markersize=10)
leg = ax3.get_legend()
leg.set_bbox_to_anchor((0.35, 0.95, 0, 0))
ax3.set_title("Courts per 10k Pink collar workers", fontsize=25)
```

### Notes on above plots

The above plots show access to public tennis courts per 10,000 white, blue, and pink collar workers in each PUMA. The economic data provided by the ACS included the total number of people in each PUMA working in management, business, service, sales, transportation, and other categories. I categorized those working in industries such as management and business as white collar workers, those working in transportation and production as blue collar workers, and those in the service industry as pink collar workers since the service industry typically involves a combination of white and blue collar work.

Based on the scale of the number of courts per 10,000 white collar workers, we can see that they have access to the fewest courts in NYC while blue and pink collar workers have similar but slightly more access. However, this is likely due simply to the very large number of white collar workers in each PUMA (mean of 45k per PUMA vs. 10k for blue collar workers and 16k for pink collar workers per PUMA). In almost every plot, we see Manhattan and lower Bronx as providing the most access to public tennis facilities.

# Conclusion regarding regressions

Out of all the  regressions, I find the above regression plot to be the most compelling because of its clarity and simplicity. The model suggests that as the percentage of poverty increases, accessibility to public tennis courts also increases, which goes against the commonly held conception (and my own assumption at the start of the project) that tennis is a sport only for the rich and therefore that residents in poor neighborhoods would have little access to tennis courts.

But  an R squared value of .18 (i.e. the model explains only 18% of the observed data) might lead one to reject any conclusions drawn from the model. Still, although 18% may sound like a very small fraction of the data, we should also consider the fact that the data analyzed in this project is highly variable and historically contingent. Decisions to build tennis courts in NYC depended on a host of factors such as the prior location of parks and the whim of  city planners. Similarly, the socioeconomic characteristics of  neighborhoods  depend greatly on historical phenomena and   human behavior that is often unpredictable.  

Given the context of the analysis, a model that explains 18% of the data actually starts to look a lot more interesting. But we cannot ignore the distribution of court access in terms of race and age. Across the city, there are more courts for white residents and residents aged 65 and older. Overall, it seems the city does well in terms of providing court access to lower income residents while older, white residents are better situated to enjoy tennis at public parks.
